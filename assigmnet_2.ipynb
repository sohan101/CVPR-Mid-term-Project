# **Initializing the process by importing the necessary tools**

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

# **Loading data from cifar10 dataset**

```python
(X_train, Y_train), (X_test, Y_test) = tf.keras.datasets.cifar10.load_data()
print(X_train.shape) #print the shape of the dataset
```

# **Training data to check the accuracy**

```python
accur = []

for k in range(1, 221):
  print("k: ", k)
  accuracy = []
  for f in range(5): 
    # 5 folds defining
    validationX, validationY = X_train[f*10000:(f+1)*10000], Y_train[f*10000:(f+1)*10000]
    trainX = np.concatenate((X_train[:f*10000], X_train[(f+1)*10000:]), axis = 0)
    trainY = np.concatenate((Y_train[:f*10000], Y_train[(f+1)*10000:]), axis = 0)
    
    cm = ce = 0      # Here, c = correct, m = manhattan, e = euclidean         

    #Range for validation declaration
    for i in range(len(validationX)):
      man_distance = []
      euc_distance = []

    #Range for trainX declaration
      for j in range(len(trainX)):
        # # manhattan (L1) - calculating 11 distance between two images
        l1 = np.sum(np.absolute(np.subtract(validationX[i], trainX[j])))

        # euclidean (L2) - calculating 12 distance between two images
        l2 = np.sqrt(np.sum((validationX[i] - trainX[j])**2))

        man_distance.append([l1, trainY[j][0]])
        euc_distance.append([l2, trainY[j][0]])
        
        if j == 400:
            break

      man_distance = np.array(man_distance)
      man_distance = man_distance[man_distance[:, 0].argsort()]
      euc_distance = np.array(euc_distance)
      euc_distance = euc_distance[euc_distance[:, 0].argsort()]
      
      # manhattan (L1) distance
      values = man_distance[:k, 1]
      most_frequent_value = np.argmax(np.bincount(values.astype('int32')))
      if most_frequent_value == validationY[i]:
        cm += 1
         # euclidean (L2) distance
      values = euc_distance[:k, 1]
      most_frequent_value = np.argmax(np.bincount(values.astype('int32')))
      if most_frequent_value == validationY[i]:
        ce += 1


      if i == 221:
        break

    accuracy.append([cm/100, ce/100])
  accur.append([accuracy, k])

print(accur)
```

# **The accuracy Graph**

```python
a1 = []
a2 = []
k = np.array(accur)[:,1]

plt.figure(figsize=(30,10))
plt.subplot(1,2,1)
plt.title('Manhattan (L1)')
for i in range(len(accur)):
  a = np.full((5), i+1, dtype=int)
  l1 = np.array(accur[i][0])[:,0] # manhattan accuracy defining
  a1.append(np.average(l1))
  plt.plot(a, l1, 'o-')
  
plt.plot(k, a1, 'o-')
plt.xlabel('k')
plt.ylabel('Accuracy of Cross-validatio')


plt.subplot(1,2,2)
plt.title('Euclidean (L2)')
for i in range(len(accur)):
  a = np.full((5), i+1, dtype=int)
  l2 = np.array(accur[i][0])[:,1] # euclidean accuracy defining
  a2.append(np.average(l2))
  plt.plot(a, l2, 'o-')

plt.plot(k, a2, 'o-')
plt.xlabel('k')
plt.ylabel('Accuracy of Cross-validation')
# Cross validation accuracy visualization between distances for k values
plt.show()
```
